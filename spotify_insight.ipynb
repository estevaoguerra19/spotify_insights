```python
import numpy as np
import pandas as pd
import seaborn as srn
```

```python
chord_highPosition = df[['Highest Charting Position','Chord']]

chord_highPosition.count()
```

```python
"""
Filtrando coluna, para facilitar a manipulação dos dados
"""

df1 = df[['Highest Charting Position','Number of Times Charted','Duration (ms)']]
df1
```

```python
"""
Conferindo se existem valores vazios
"""

duration = df['Duration (ms)']

duration.where(duration == " ").count()
```

```python
"""
Substituindo valores vazios parar 0
"""

duration = duration.replace(" ", 0)
duration
```

```python
duration
```

```python
"""
Renomeando a coluna 'Duration (ms)' para 'Duration (min)'
"""
df = df.rename(columns = {'Duration (ms)' : 'Duration (min)'})

df
```

```python
"""
Transformando os milímetros em segundos
"""

duration = pd.to_numeric(duration, downcast='float')

duration =  (duration / 1000) / 60

duration = round(duration, 2)
duration
```
